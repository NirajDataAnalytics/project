```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df=df.set_index('Formatted Date')
Weather_Data=df[['Apparent Temperature (C)','Visibility (km)','Humidity','Wind Speed (km/h)']].resample('MS').mean()
print(Weather_Data)
```

```python
plt.figure(figsize=(15,5))
plt.plot(Weather_Data['Humidity'], label='Humidity', color='red',linestyle='dashed')
plt.plot(Weather_Data['Apparent Temperature (C)'], label='App Temp', color='blue')
plt.title('Varitaion of Apparent Temperature Vs Humidity', fontsize=10)
plt.legend(loc = 'lower right', fontsize = 12)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Year')
plt.ylabel('Numbers')
plt.grid(color='black', linestyle='--', linewidth=0.8)
```

```python
df1=Weather_Data.reset_index()
df1['Formatted Date']=pd.to_datetime(df1['Formatted Date'], utc=True)
df1['month']=df1['Formatted Date'].dt.month
df1['year']=df1['Formatted Date'].dt.year
```

```python
Data_1 = df1.groupby(['month','year']).mean()
print(Data_1)
```

```python
plt.figure(figsize=(15,5))
range_years=np.arange(2006,2017)
plt.plot(range_years, Data_1['Humidity'][121:132].values, marker='*', color='green', label='Dec Humidity')
plt.plot(range_years, Data_1['Apparent Temperature (C)'][121:132].values, marker='o', color='green', label='Dec Temp')
plt.plot(range_years, Data_1['Humidity'][33:44].values, marker='*', color='yellow', label='April Humidity')
plt.plot(range_years, Data_1['Apparent Temperature (C)'][33:44].values, marker='o', color='yellow', label='April Temp')
plt.title('Variation of Humidity and Temp in the Month of April and Dec from 2006 to 2016', fontsize=10)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.legend('centre right', fontsize = 10)
plt.xlabel('Year')
plt.ylabel('Values')
plt.grid(color='black', linestyle='--', linewidth= 0.25)
```

```python
correlation=Weather_Data.corr()
sns.heatmap(correlation, annot=True,fmt='.2%')
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Apparent Temperature (C)', y= 'Humidity', data=Weather_Data[Weather_Data.index.month==4])
plt.xticks(rotation=90)
plt.title('Relation between temperature and humidity for the month of April')
plt.show()
```

```python
plt.figure(figsize=(15,3));
plt.plot(Weather_Data['Wind Speed (km/h)'], label = 'Wind Speed (km/h)', color ='red',linestyle='dotted')
plt.plot(Weather_Data['Apparent Temperature (C)'], label = 'App Temp (oC)', color  = 'blue', linestyle='dashed')
plt.title('Variation of Wind Speed (km/h) v/s App Temp (oC)', fontsize = 10)
plt.legend(loc=0, fontsize =12)
plt.xticks(fontsize = 15)
plt.yticks(fontsize=13)
plt.xlabel('Year')
plt.ylabel('Values')
plt.grid(color='black', linestyle='--', linewidth= 0.25)
```

```python
may=Weather_Data[Weather_Data.index.month==5]
plt.figure(figsize=(15,3))
plt.plot(may.loc['2006-05-01':'2016-05-01','Apparent Temperature (C)'], marker='o', label='App Temp (0C)', color= 'green')
plt.plot(may.loc['2006-05-01':'2016-05-01','Wind Speed (km/h)'],marker='o', label = 'Wind Speed (km/h)', color ='red',linestyle='dotted')
plt.plot(may.loc['2006-05-01':'2016-05-01','Visibility (km)'],marker='o', label = 'Visibility (km)', color ='orange',linestyle='dotted')
plt.legend(loc='best', fontsize =10)
plt.xlabel('Month of May',fontsize = 10)
plt.ylabel('Values')
plt.grid(color='black',linestyle='--', linewidth= 0.25)
plt.title('Wind Speed (km/h), Visibility (km) and Apparent Temperature', fontsize = 10)
```

```python
df_num=df.select_dtypes(exclude=object)
df_cat=df.select_dtypes(include=object)
```

```python
plt.figure(figsize=(20,5))

re=1
for i in df_num.columns:
    plt.subplot(3,3,re)
    sns.boxplot(df[i], palette='RdPu')
    plt.title(i)
    re+=1
plt.tight_layout()
```

```python
plt.figure(figsize=(20,20))

re=1
for i in df_num.columns:
    plt.subplot(3,3,re)
    sns.kdeplot(df[i], shade=True,color='orange',cbar=True)
    plt.title(i)
    re+=1
plt.tight_layout()
```

```python
plt.figure(figsize=(20,10))
sns.kdeplot(
    data=df,
    x="Humidity",
    y="Temperature (C)",
    hue="Precip Type",
    thresh=.1)
plt.title('Multiple bivariate KDE plot')
```
